## Gradient Descent algorithm

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline 

import sklearn as sns
```

```python
x = np.array([34,108,64,88,99,51])
y = np.array([5,17,11,8,14,5])
```

```python
# With the known features and target variables creating a CSV file
import csv
with open('Gradient_Descent.csv','w',newline='\n') as file:
    writer = csv.writer(file)
    writer.writerow(['Bill amount','Tip amount'])
    writer.writerow([34,5])
    writer.writerow([108,17])
    writer.writerow([64,11])
    writer.writerow([88,8])
    writer.writerow([99,14])
```

```python
import csv
with open('Gradient_Descent(New).csv','w',newline='\n') as file:
    writer = csv.writer(file)
    writer.writerow(['Bill amount','Tip amount'])
    for i,j in zip(x,y):
        writer.writerow([i,j])
```

```python
df=pd.read_csv("Gradient_Descent(New).csv")

x = df[['Bill amount']]
y = df[['Tip amount']]
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(x,y)
```

```python
print("Coefficient : ",lm.coef_)
print("Intercept   : ",lm.intercept_)
```

```python
# Then the equation is y = (-0.33949373)*x + 58.48420699

plt.xlabel('Bill Amount(US $)')
plt.ylabel('Tips(US $)')
plt.scatter(x,y,color="red",marker="*")

plt.plot()
```

```python
lm.predict(x)
```

```python
def Gradient_Descent(x,y,iterations,learning_rate):
    m=c=0
    n=len(x)   # Number of rows in the data-set
    
#    J = 1/n { summation(from i=1 to n) [y(observed) - y(predict)]^2 }
    for i in range(iterations):
        y_predict = (m*x) + c
        J  = (1/n)*sum([val**2 for val in (y-y_predict)])  # Cost function
        Jm = -(2/n)*sum( x*(y-y_predict))  # Partial derivative of J with respect to m
        Jc = -(2/n)*sum(y-y_predict)      # Partial derivative of J with respect to c
         
        m = m-(learning_rate * Jm)
        c = c-(learning_rate * Jc)
        
        print("m: {}, c: {}, J: {}, iterations: {}".format(m,c,J,i))
```

```python
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
Gradient_Descent(x,y,1000,0.01)
```
