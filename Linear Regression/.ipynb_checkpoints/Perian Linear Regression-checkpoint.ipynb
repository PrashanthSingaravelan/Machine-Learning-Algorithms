```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv("USA_Housing.csv")
```

```python
# For the purpose of splitting the data-set

# Feature variable
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]

# Target variable
y = df['Price']
```

```python
# Spliting into Training and Testing data-set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=101)
```

```python
print("X_train.shape  : ",X_train.shape) # 2000 rows    (2/5)*100 = 40%
print("X_test.shape   : ",X_test.shape)  # 3000 rows    (3/5)*100 = 60%
print("Data-set shape : ",df.shape)      # 5000 rows    Total = 100%

# From the above, it is clear that the complete data-set has benn divided into 
# Training data-set : 40%
# Testing data-set  : 60%
```

```python
# Take the linear regression model
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

# After splitting the data-set ,take the training data-set and fir it into our model
lm=lm.fit(X_train,y_train)
```

```python
lm.intercept_        # intercept (c=-2640159.7968132403)
```

```python
lm.coef_ # slope (m=[2.15282755e+01, 1.64883282e+05, 1.22368678e+05, 2.23380186e+03,1.51504200e+01])
```

```python
cdf = pd.DataFrame(lm.coef_ ,X_train.columns,columns=["Coefficient"])
cdf

# For every 1-unit inc of Avg.Area Income           inc 21.528276 dollars in price
# For every 1-unit inc of Avg.Area House Age        inc 164883.282027 dollars in price
# For every 1-unit inc of Avg. Area Number of Rooms inc 122368.678023 dollars in price
# For every 1-unit inc of Avg. Area Number of Bedrooms inc 2233.801864 dollars in price
# For every 1-unit inc of Area Populatione          inc 15.150420 dollars in price

```

## Predictions

```python
# y= mx+c
# slope (m=[2.15282755e+01, 1.64883282e+05, 1.22368678e+05, 2.23380186e+03,1.51504200e+01])
# intercept (c=-2640159.7968132403)

y_predict = lm.predict(X_test)
print("Predicted Y value : ",y_predict)
print()
print("Actual Y value    : ",y_test)
```

```python
# Analysing the Predicted and Observed data-points
plt.scatter(y_test,y_predict)

# Looks like a perfect predictions and somewhat looks like a perfect straight line, then it is perfect for Linear regression
```

```python
## Visualising the residual errors
## residual error = Actual-Observed
sns.distplot(y_test-y_predict)
```

```python
# If the above residual model is normally distribured, then the linear regression model is preferred for the data-set
# If the graph is not normally distributed, then linear regression is not the preferred one for the data-set
```

```python
from sklearn import metrics

print(metrics.mean_absolute_error(y_test,y_predict))

print(metrics.mean_squared_error(y_test,y_predict))

# Root MeanSquared Error
print(np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```
