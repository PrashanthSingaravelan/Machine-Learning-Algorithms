```python
import numpy as np
import sklearn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Gradient_Descent(x,y,iterations,learning_rate):
    m=c=0
    n=len(x)   # Number of rows in the data-set
    
#    J = 1/n { summation(from i=1 to n) [y(observed) - y(predict)]^2 }
    for i in range(iterations):
        y_predict = (m*x) + c
        
        J  = (1/n)*sum([val**2 for val in (y-y_predict)])  # Cost function
        Jm = -(2/n)*sum( x*(y-y_predict))  # Partial derivative of J with respect to m
        Jc = -(2/n)*sum(y-y_predict)       # Partial derivative of J with respect to c
         
        m = m-(learning_rate * Jm)   # learning rate --> will tell us how much jump we are undergoing
        c = c-(learning_rate * Jc)
        
        print("m: {}, c: {}, J: {}, iterations: {}".format(m,c,J,i))
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

### Main goal :  Each and every time if the cost function is reduced, then the algorithm is correct

```python
# By increasing the number of iterations the accuracy is too close
Gradient_Descent(x,y,1000,0.001)
# J = 0.47797941992396514
```

```python
# By reducing the number of iterations the cost function is increased
Gradient_Descent(x,y,100,0.001)
# J = 1.6469972426658057

# Hence J depends upon the number of iterations
```

```python
# By reducing the learning rate also the cost function increases
Gradient_Descent(x,y,100,0.1)
# J: 5.970406709462973e+28
```
